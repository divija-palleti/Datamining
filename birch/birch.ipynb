```python
T = 0
L = 0
B = 0
data_points = []

with open('input.txt', 'r') as f:
    data = f.readlines()
    (T, L, B) = list(map(float, data[0].split(' ')))
    data_points = list(map(float, data[1:]))


class Node:
    def __init__(self):
        self.n = 0
        self.ls = 0
        self.ss = 0
        self.children = []
        self.parent = None

    def insert_point(self, value):
        is_inserted = False
        sorted_children = sorted(self.children, key=lambda x: abs((x.ls / x.n) - value))

        for child in sorted_children:
            if child.insert_point(value):
                is_inserted = True

                break

        if not is_inserted and len(self.children) < B:
            if len(self.children) == 0 or isinstance(self.children[0], Leaf):
                new_leaf = Leaf()
                new_leaf.parent = self
                self.children.append(new_leaf)
                new_leaf.insert_point(value)
            elif isinstance(self.children[0], Node):
                new_node = Node()
                new_node.parent = self
                self.children.append(new_node)
                new_node.insert_point(value)

            return True

        if not is_inserted and len(self.children) >= B:
            if self.parent != None:
                new_node = Node()

                self.parent.children.append(new_node)
                self.parent.children.remove(self)
                new_node.parent = self.parent
                self.parent = new_node
                new_node.insert_point(value)

                new_node.children.append(self)

                return True
            else:
                return False

        return is_inserted

    def update_atributes(self):
        self.n = 0
        self.ls = 0
        self.ss = 0

        for child in self.children:
            self.n += child.n
            self.ls += child.ls
            self.ss += child.ss

        if self.parent != None:
            self.parent.update_atributes()


class Leaf:
    def __init__(self):
        self.n = 0
        self.ls = 0
        self.ss = 0
        self.points = []
        self.parent = None
        self.radius = 0

    def new_radius(self, point):
        new_n = self.n + 1
        new_ls = self.ls + point
        new_ss = self.ss + point ** 2
        new_radius = ((new_n * new_ss - (new_ls ** 2)) / (new_n ** 2)) ** 0.5

        return new_radius

    def can_insert(self, point):
        if self.n >= L:
            return False

        return self.new_radius(point) <= T

    def insert_point(self, point):
        if not self.can_insert(point):
            return False

        self.points.append(point)
        new_n = self.n + 1
        new_ls = self.ls + point
        new_ss = self.ss + point ** 2
        new_radius = ((new_n * new_ss - (new_ls ** 2)) / (new_n ** 2)) ** 0.5
        self.n = new_n
        self.ls = new_ls
        self.ss = new_ss
        self.radius = new_radius

        self.parent.update_atributes()

        return True


def print_tree(root):
    if isinstance(root, Node):
        print(root.n, root.ls, root.ss)

        for child in root.children:
            print_tree(child)
    else:
        print(root.n, root.ls, root.ss, root.radius)


root = Node()

for val in data_points:
    is_inserted = root.insert_point(val)

    if not is_inserted:
        new_root = Node()
        new_root.children.append(root)
        #root.parent = new_root
        temp = root
        root = new_root
        print(root.children)
        print(root.parent)
        #if val != 0.65:
        root.insert_point(val)
        temp.parent = root

    print_tree(root)
    print('--------------------------------')

print_tree(root)
```
